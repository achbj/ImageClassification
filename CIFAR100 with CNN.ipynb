## 1 Import

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim


import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader

import os
import numpy as np
import matplotlib.pyplot as plt

from tqdm import tqdm
```

```python
# reproducibility
torch.manual_seed(42)
np.random.seed(42)
```

## Params

```python
config = {
    "data_root": "./data",
    "num_classes": 100,
    "batch_size": 64,
    "num_workers": 2,
    "epochs": 60,                    # longer training
    "momentum": 0.9,
    "weight_decay": 5e-4,
    "learning_rate": 0.05,           # much higher initial LR
    "input_channels": 3,
    "conv_kernel_size": 3,
    "conv_padding": 1,               # new key
    # filter progression per conv layer
    "conv_channels": [64, 128, 256], # three layers
    "pool_kernel_size": 2,
    "pool_stride": 2,
    "fc1_in_features": 256 * 4 * 4,  # after three poolings: 32->16->8->4
    "fc1_out_features": 256,         # larger hidden FC
    "dropout1_rate": 0.25,
    "dropout2_rate": 0.5,
    "save_path": "./best_model.pt"
}
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

## Data Transformation

```python
# Cifar normalization values provided in pytorch documentation
# mean = (0.5071, 0.4867, 0.4408)
# std = (0.2675, 0.2565, 0.2761)

# Training set with data augmentation
train_transform = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=(0.5071, 0.4867, 0.4408),
                         std=(0.2675, 0.2565, 0.2761)),
])

# Validation / Test set with no augmentation
test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=(0.5071, 0.4867, 0.4408),
                         std=(0.2675, 0.2565, 0.2761)),
])
```

## Download Data & Create Object

```python
train_dataset = torchvision.datasets.CIFAR100( root=config["data_root"],
                                              train=True, download=True,
                                              transform=train_transform
                                             )
test_dataset = torchvision.datasets.CIFAR100(root=config["data_root"],
                                             train=False, download=True,
                                             transform=test_transform
                                            )
# Split a validation set from the training set 10%
val_size = int(0.1 * len(train_dataset))
train_size = len(train_dataset) - val_size
train_subset, val_subset = torch.utils.data.random_split(
    train_dataset, [train_size, val_size]
)
```

## Create Loader

```python
train_loader = DataLoader(
    train_subset,
    batch_size=config["batch_size"],
    shuffle=True,
    num_workers=config["num_workers"],
    pin_memory=True,
)
val_loader = DataLoader(
    val_subset,
    batch_size=config["batch_size"],
    shuffle=False,
    num_workers=config["num_workers"],
    pin_memory=True,
)
test_loader = DataLoader(
    test_dataset,
    batch_size=config["batch_size"],
    shuffle=False,
    num_workers=config["num_workers"],
    pin_memory=True,
)
```

```python
class SimpleCNN(nn.Module):
    def __init__(self, config):
        super(SimpleCNN, self).__init__()
        ch = config["conv_channels"]  # list of output channels
        pad = config["conv_padding"]
        ks = config["conv_kernel_size"]

        # Convolutional blocks
        self.conv1 = nn.Conv2d(config["input_channels"], ch[0], ks, padding=pad)
        self.bn1   = nn.BatchNorm2d(ch[0])
        self.pool1 = nn.MaxPool2d(config["pool_kernel_size"], config["pool_stride"])
        self.drop1 = nn.Dropout(config["dropout1_rate"])

        self.conv2 = nn.Conv2d(ch[0], ch[1], ks, padding=pad)
        self.bn2   = nn.BatchNorm2d(ch[1])
        self.pool2 = nn.MaxPool2d(config["pool_kernel_size"], config["pool_stride"])
        self.drop2 = nn.Dropout(config["dropout1_rate"])

        self.conv3 = nn.Conv2d(ch[1], ch[2], ks, padding=pad)
        self.bn3   = nn.BatchNorm2d(ch[2])
        self.pool3 = nn.MaxPool2d(config["pool_kernel_size"], config["pool_stride"])
        self.drop3 = nn.Dropout(config["dropout2_rate"])   # higher dropout before FC

        # Fully‑connected
        self.fc1 = nn.Linear(config["fc1_in_features"], config["fc1_out_features"])
        self.drop_fc = nn.Dropout(config["dropout2_rate"])
        self.fc2 = nn.Linear(config["fc1_out_features"], config["num_classes"])

    def forward(self, x):
        x = F.relu(self.bn1(self.conv1(x)))
        x = self.drop1(self.pool1(x))

        x = F.relu(self.bn2(self.conv2(x)))
        x = self.drop2(self.pool2(x))

        x = F.relu(self.bn3(self.conv3(x)))
        x = self.drop3(self.pool3(x))

        x = torch.flatten(x, start_dim=1)
        x = F.relu(self.fc1(x))
        x = self.drop_fc(x)
        x = self.fc2(x)
        return x
```

```python
# initalize
model = SimpleCNN(config).to(device)
print(model)
```

### Loss, Optimizer, Scheduler

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(
    model.parameters(),
    lr=config["learning_rate"],
    momentum=config["momentum"],
    weight_decay=config["weight_decay"],
)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=config["epochs"])
```

```python
def train_one_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    loop = tqdm(loader, desc="Training", leave=False)
    for images, labels in loop:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        loop.set_postfix(loss=loss.item(), acc=100.0 * correct / total)

    epoch_loss = running_loss / total
    epoch_acc = 100.0 * correct / total
    return epoch_loss, epoch_acc
```

```python
@torch.no_grad()
def evaluate(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    loop = tqdm(loader, desc="Evaluating", leave=False)
    for images, labels in loop:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        loss = criterion(outputs, labels)

        running_loss += loss.item() * images.size(0)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        loop.set_postfix(loss=loss.item(), acc=100.0 * correct / total)

    epoch_loss = running_loss / total
    epoch_acc = 100.0 * correct / total
    return epoch_loss, epoch_acc
```

```python
history = {
    "train_loss": [],
    "train_acc": [],
    "val_loss": [],
    "val_acc": [],
}
```

```python
best_val_acc = 0.0
for epoch in range(1, config["epochs"] + 1):
    print(f"\nEpoch {epoch}/{config['epochs']}")

    train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = evaluate(model, val_loader, criterion, device)

    scheduler.step()

    history["train_loss"].append(train_loss)
    history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss)
    history["val_acc"].append(val_acc)

    print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
    print(f"Val   Loss: {val_loss:.4f} | Val   Acc: {val_acc:.2f}%")

    # Save best model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), config["save_path"])
        print(f"  -> Saved new best model with Val Acc: {val_acc:.2f}%")

print(f"\nTraining complete. Best validation accuracy: {best_val_acc:.2f}%")
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(history["train_loss"], label="Train Loss")
ax1.plot(history["val_loss"], label="Validation Loss")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.set_title("Loss Curves")
ax1.legend()

ax2.plot(history["train_acc"], label="Train Accuracy")
ax2.plot(history["val_acc"], label="Validation Accuracy")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Accuracy (%)")
ax2.set_title("Accuracy Curves")
ax2.legend()

plt.tight_layout()
plt.savefig("training_curves.png", dpi=150)
plt.show()
```
